This is dataset preparation notebook for Text Generation training in LLaMA.

```python
import pandas as pd
import json
import re
```

```python
train_meta=pd.read_csv(r'DAIC\train_split_Depression_AVEC2017.csv')
test_meta=pd.read_csv(r'DAIC\full_test_split.csv')
val_meta=pd.read_csv(r'DAIC\dev_split_Depression_AVEC2017.csv')
```

```python
data=[]
for id in train_meta['Participant_ID']:
    file_dir=dir+str(id)+'_TRANSCRIPT.csv'
    file=pd.read_csv(file_dir,sep='\t')
    label=train_meta['PHQ8_Binary'].loc[train_meta['Participant_ID']==id].item()
    ds=[
    {'role':"system",
     'content':"You are an assistant. Your job is to ask relevant questions to probe deeper into user's emotional state"}
    ]
    previous_speaker=file['speaker'][0]
    content=''
    for ind in file.index:
        try:
            current_speaker=file['speaker'][ind]
            if current_speaker==previous_speaker:
                file['value'][ind]=file['value'][ind].replace('<sync>','')
                file['value'][ind]=file['value'][ind].replace('<synch>','')
                file['value'][ind]=file['value'][ind].replace('(','')
                file['value'][ind]=file['value'][ind].replace(')','')
                file['value'][ind]=file['value'][ind].replace('  ',' ')
                file['value'][ind]=re.sub(r"<(\w*)>",r"\1",file['value'][ind])
                content=content+' '+file['value'][ind]
                continue
            else:
                if previous_speaker=='Ellie':
                    role_val='assistant'
                else:
                    role_val='user'
                datum={'role':role_val,
                   'content':content}
                ds.append(datum)
                previous_speaker=current_speaker
                file['value'][ind]=file['value'][ind].replace('<sync>','')
                file['value'][ind]=file['value'][ind].replace('<synch>','')
                file['value'][ind]=file['value'][ind].replace('(','')
                file['value'][ind]=file['value'][ind].replace(')','')
                file['value'][ind]=file['value'][ind].replace('  ',' ')
                file['value'][ind]=re.sub(r"<(\w*)>",r"\1",file['value'][ind])
                content=file['value'][ind]
        except:
            print(id,ind,file['value'][ind])
    datum={'role':role_val,
                   'content':content}
    ds.append(datum)
    data.append({'data':ds})

with open('train_instruct.json', 'w') as f:
    json.dump(data, f, indent=4)
```
