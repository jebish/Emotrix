```python
#Training but balancing the dataset
data=[]
total0=0
for id in train_meta['Participant_ID']:
    file_dir=dir+str(id)+'_TRANSCRIPT.csv'
    file=pd.read_csv(file_dir,sep='\t')
    label=train_meta['PHQ8_Binary'].loc[train_meta['Participant_ID']==id].item()
    if label==0:
        total0+=1
        if total0>30:
            continue
    ds=[
    {'role':"system",
     'content':"You are a assistant. Below is an instruction with a conversation that describes a task. Write a response that appropriately completes the task"}
    ]
    
    datum={
    'instruction':'''Analyze the conversation  and Categorize the Participant as one of these classes: 
    Depressed
    Not Depressed
    Don't output anything else and use Ellie's questions for understanding conversation flow
    ''',
    'input':'',
    }
    for ind in file.index:
        try:
            file['value'][ind]=file['value'][ind].replace('<sync>','')
            file['value'][ind]=file['value'][ind].replace('<synch>','')
            file['value'][ind]=file['value'][ind].replace('  ',' ')
            file['value'][ind]=re.sub(r"<(\w*)>",r"\1",file['value'][ind])
            datum['input']+=file['speaker'][ind]+': '+file['value'][ind]+'\n'
        except:
            print(id,ind,file['value'][ind])
    datum['input']=datum['input'].rstrip('\n')
    datum='Instruction: '+datum['instruction']+'\n Input: '+datum['input']
    dat={
        'role':'user',
        'content':datum
    }
    ds.append(dat)
    out={
        'role':'assistant',
        'content':'Depressed' if label==1 else 'Not Depresssed'
    }
    ds.append(out)
    data.append({'data':ds})

with open('train_data_bal.json', 'w') as f:
    json.dump(data, f, indent=4)
    
```
